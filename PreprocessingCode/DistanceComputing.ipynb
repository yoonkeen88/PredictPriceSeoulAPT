## 거리 계산
- 부동산 실거래 데이터에서 대학교, 역, 대형 문화시설 등과의 거리를 계산하여 컬럼 추가
- 거리 계산 후에는 좌표 컬럼은 제거

#### 필요한 패키지 설치

```python
import math
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
# 서울만을 다루는 거리이기에 유클리드 거리로 계산

def euclidean_distance(lat1, lon1, lat2, lon2):
    # 위도와 경도를 미터로 변환
    lat1_rad = math.radians(lat1)
    lon1_rad = math.radians(lon1)
    lat2_rad = math.radians(lat2)
    lon2_rad = math.radians(lon2)

    # 위도와 경도 차이 계산 (미터로 변환)
    lat_diff = (lat2_rad - lat1_rad) * 6371000  # 위도 차이 (미터 단위)
    lon_diff = (lon2_rad - lon1_rad) * 6371000 * math.cos((lat1_rad + lat2_rad) / 2)  # 경도 차이 (미터 단위)

    # 유클리드 거리 계산 (미터 단위)
    distance = math.sqrt(lat_diff ** 2 + lon_diff ** 2)

    return distance
```
